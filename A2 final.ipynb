````python
import torch
import random
import numpy as np
import os
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset
from collections import defaultdict

# Set random seed for reproducibility
torch.manual_seed(999)
random.seed(999)  # Ensures Python’s random module is deterministic
np.random.seed(999)  # Ensures NumPy is deterministic


def get_data_loader(data_dir, batch_size, valid_split=0.2, test_split=0.1):
    """
    Load dataset while ensuring test, train, and validation sets contain completely unique individuals.

    Test and validation sets only include individuals with all 9 letters (A-I), each having exactly 3 images.
    Training set includes both valid and incomplete individuals.

    Args:
        data_dir: The root directory of the dataset, with subdirectories for each letter class (A-I).
        batch_size: Number of samples per batch.
        valid_split: Fraction of remaining people (after test split) used for validation.
        test_split: Fraction of total people used for testing.

    Returns:
        train_loader, val_loader, test_loader, classes
    """

    # Define image transformations
    transform = transforms.Compose([
        transforms.Resize((224, 224)),  # Resize images to 224x224
        transforms.ToTensor(),          # Convert image to tensor
        transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalize RGB channels
    ])

    # Load the dataset
    dataset = datasets.ImageFolder(data_dir, transform=transform)
    classes = dataset.classes  # List of class names (A-I)

    # Step 1: Group images by person ID and letter
    person_to_letters = defaultdict(lambda: defaultdict(list))

    for idx, (img_path, label) in enumerate(dataset.imgs):
        filename = os.path.basename(img_path)  # Extract filename (e.g., "1_G_1.jpg")
        parts = filename.split("_")
        if len(parts) == 3:
            person_id, letter, seq = parts
            person_to_letters[person_id][letter].append(idx)  # Store image index for this letter/person

    # Step 2: Separate valid and invalid individuals
    valid_people = {person: images for person, images in person_to_letters.items() if len(images) == 9 and all(len(images[letter]) == 3 for letter in images)}
    invalid_people = {person: images for person, images in person_to_letters.items() if person not in valid_people}

    # Get total number of people
    total_people = len(person_to_letters)

    # Step 3: Shuffle valid individuals
    valid_people_list = list(valid_people.keys())
    random.shuffle(valid_people_list)

    # Step 4: Compute test and validation sizes
    test_people_count = round(test_split * total_people)  # 10% of total people
    remaining_valid_count = len(valid_people_list) - test_people_count  # Remaining valid people after test selection
    valid_people_count = round((total_people - test_people_count) * valid_split)  # 20% of the remaining 90%

    # Step 5: Assign test and validation sets from valid people
    test_people = set(valid_people_list[:test_people_count])  # First 10% of total people
    train_valid_people = valid_people_list[test_people_count:]  # Remaining valid people
    val_people = set(train_valid_people[:valid_people_count])  # 20% of the remaining 90%
    train_people = set(train_valid_people[valid_people_count:])  # Remaining valid people go to training

    # Step 6: Add incomplete individuals to training set
    train_people.update(invalid_people.keys())  # Training set gets all remaining people

    # Step 7: Collect image indices
    test_indices = [idx for person in test_people for letter in valid_people[person].values() for idx in letter]
    val_indices = [idx for person in val_people for letter in valid_people[person].values() for idx in letter]
    train_indices = [idx for person in train_people for letter in person_to_letters[person].values() for idx in letter]

    # # Step 8: Print selected individuals with details
    # print("Test set individuals and their letters:")
    # for person in test_people:
    #     print(f"Person {person}:")
    #     for letter, indices in valid_people[person].items():
    #         print(f"  Letter {letter}: {indices}")

    # print("\nValidation set individuals and their letters:")
    # for person in val_people:
    #     print(f"Person {person}:")
    #     for letter, indices in valid_people[person].items():
    #         print(f"  Letter {letter}: {indices}")

    # Step 9: Create DataLoaders
    train_loader = DataLoader(Subset(dataset, train_indices), batch_size=batch_size, shuffle=False)
    val_loader = DataLoader(Subset(dataset, val_indices), batch_size=batch_size, shuffle=False)
    test_loader = DataLoader(Subset(dataset, test_indices), batch_size=batch_size, shuffle=False)

    # Step 10: Verify correct number of people in each set
    # Get the number of images in each split
    num_train = len(train_loader.dataset)
    num_val = len(val_loader.dataset)
    num_test = len(test_loader.dataset)
    print(f"\nTotal images: {num_train + num_val + num_test}")
    print(f"Training images: {num_train}")
    print(f"Validation images: {num_val}")
    print(f"Test images: {num_test}\n")
    print(f"Train set contains {len(train_people)} unique individuals.")
    print(f"Validation set contains {len(val_people)} unique individuals.")
    print(f"Test set contains {len(test_people)} unique individuals.\n")

    return train_loader, val_loader, test_loader, classes
````

### Part (i) Convolutional Network [2pt MODEL]

Build a convolutional neural network model that takes the (224x224 RGB) image as input, and predicts the gesture
letter. Your model should be a subclass of nn.Module. Explain your choice of neural network architecture: how
many layers did you choose? What types of layers did you use? Were they fully-connected or convolutional?
What about other decisions like pooling layers, activation functions, number of channels / hidden units?

````python
import torch
import torch.nn as nn
import torch.nn.functional as F

class GestureCNN(nn.Module):
    def __init__(self, conv1_channels=16, conv2_channels=32, conv3_channels=64,
                 fc1_units=128, kernel=5, stride=1, padding=0, input_size = 224):
        super(GestureCNN, self).__init__()
        self.name = "GCNN"

        # Define convolutional layers with configurable stride and padding
        self.conv1 = nn.Conv2d(3, conv1_channels, kernel, stride=stride, padding=padding)
        self.conv2 = nn.Conv2d(conv1_channels, conv2_channels, kernel, stride=stride, padding=padding)
        self.conv3 = nn.Conv2d(conv2_channels, conv3_channels, kernel, stride=stride, padding=padding)

        # Max pooling layer (kernel size = 2, stride = 2)
        self.pool = nn.MaxPool2d(2, 2)

        # Compute feature map size after three convolutional + pooling layers
        size = input_size  # Input image size (224x224)
        for _ in range(3):
            size = (size - kernel + 2 * padding) // stride + 1  # Convolution output
            size = size // 2  # Pooling output

        # Fully connected layers
        self.fc1 = nn.Linear(conv3_channels * size * size, fc1_units)  # Adjusted FC input size
        self.fc2 = nn.Linear(fc1_units, 9)  # 9 output classes

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))  # Conv1 -> ReLU -> Pooling
        x = self.pool(F.relu(self.conv2(x)))  # Conv2 -> ReLU -> Pooling
        x = self.pool(F.relu(self.conv3(x)))  # Conv3 -> ReLU -> Pooling

        x = x.view(x.size(0), -1)  # Flatten feature map for fully connected layers
        x = F.relu(self.fc1(x))  # Fully connected layer with ReLU activation
        x = self.fc2(x)  # Output layer (logits)

        return x
````

### Part (ii) Training Code [1pt MODEL]

Write code that trains your neural network given some training data. Your training code should make it easy
to tweak the usual hyperparameters, like batch size, learning rate, and the model object itself. Make sure
that you are checkpointing your models from time to time (the frequency is up to you). Explain your choice
of loss function and optimizer.

````python
import os
import time
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm

def train_model(model,
                dataset_path="dataset",
                batch_size=32,
                num_epochs=10,
                lr=0.001,
                model_path="checkpoints/model",
                get_data_loader_fn=None,
                optimizer_fn=optim.Adam,
                criterion_fn=nn.CrossEntropyLoss,
                conv1_channels=16,
                conv2_channels=32,
                conv3_channels=64,
                fc1_units=128,
                kernel=5,
                stride=1,
                padding=0):
    """
    Trains the GestureCNN model with adjustable hyperparameters including batch size, learning rate, optimizer, loss function, and model architecture.

    Args:
        model (nn.Module): The CNN model to train.
        dataset_path (str): Path to the dataset directory.
        batch_size (int): Batch size for training and validation.
        num_epochs (int): Number of training epochs.
        lr (float): Learning rate.
        model_path (str): Path prefix for saving model checkpoints and logs.
        get_data_loader_fn (function): Function to load dataset.
        optimizer_fn (function): Optimizer function (default: Adam).
        criterion_fn (function): Loss function (default: CrossEntropyLoss).
        conv1_channels (int): Number of channels in the first convolutional layer.
        conv2_channels (int): Number of channels in the second convolutional layer.
        conv3_channels (int): Number of channels in the third convolutional layer.
        fc1_units (int): Number of hidden units in the fully connected layer.
        kernel (int): Kernel size for convolutional layers.
        stride (int): Stride value for convolutional layers.
        padding (int): Padding value for convolutional layers.

    Returns:
        model (nn.Module): The trained model.
    """
    model = model(
        conv1_channels=conv1_channels,
        conv2_channels=conv2_channels,
        conv3_channels=conv3_channels,
        fc1_units=fc1_units,
        kernel=kernel,
        stride=stride,
        padding=padding
    )

    # Load dataset using the provided function
    if get_data_loader_fn is None:
        raise ValueError("get_data_loader_fn must be provided to load the dataset.")

    train_loader, val_loader, test_loader, classes = get_data_loader_fn(dataset_path, batch_size=batch_size)

    # Set device (GPU if available, otherwise CPU)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # Define loss function and optimizer
    criterion = criterion_fn()
    optimizer = optimizer_fn(model.parameters(), lr=lr)

    # Variables to track the best model
    best_val_acc = 0.0
    best_val_loss = float("inf")
    best_model_path = ""

    # Create directory for saving checkpoints
    checkpoint_dir = os.path.dirname(model_path)
    os.makedirs(checkpoint_dir, exist_ok=True)

    # Lists to store training and validation metrics
    train_loss_list, train_acc_list = [], []
    val_loss_list, val_acc_list = [], []

    # Start training time tracking
    start_time = time.time()

    for epoch in range(num_epochs):
        model.train()
        running_loss, correct, total = 0.0, 0, 0

        # Training loop
        for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}"):
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()  # Reset gradients
            outputs = model(images)  # Forward pass
            loss = criterion(outputs, labels)  # Compute loss
            loss.backward()  # Backpropagation
            optimizer.step()  # Update model parameters

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

        # Compute training loss and accuracy
        train_loss = running_loss / len(train_loader)
        train_accuracy = correct / total
        train_loss_list.append(train_loss)
        train_acc_list.append(train_accuracy)

        if val_loader is None:
          print(f"Epoch {epoch+1}/{num_epochs} - Train Loss: {train_loss:.4f}, Train Acc: {train_accuracy * 100:.4f}%")
        else:
          # Validation loop
          model.eval()
          val_loss, val_correct, val_total = 0.0, 0, 0
          with torch.no_grad():
              for images, labels in val_loader:
                  images, labels = images.to(device), labels.to(device)
                  outputs = model(images)
                  loss = criterion(outputs, labels)
                  val_loss += loss.item()
                  _, predicted = outputs.max(1)
                  val_correct += (predicted == labels).sum().item()
                  val_total += labels.size(0)

          # Compute validation loss and accuracy
          val_loss /= len(val_loader)
          val_accuracy = val_correct / val_total
          val_loss_list.append(val_loss)
          val_acc_list.append(val_accuracy)

          print(f"Epoch {epoch+1}/{num_epochs} - Train Loss: {train_loss:.4f}, Train Acc: {train_accuracy * 100:.4f}%, "
                f"Val Loss: {val_loss:.4f}, Val Acc: {val_accuracy * 100:.4f}%")

          # Save checkpoint for each epoch
          checkpoint_epoch_path = f"{model_path}_checkpoint_epoch_{epoch+1}.pth"
          torch.save(model.state_dict(), checkpoint_epoch_path)
          print(f"Checkpoint saved at {checkpoint_epoch_path}")

          # Save the best model based on validation accuracy
          val_acc_rounded = round(val_accuracy, 2)
          if val_acc_rounded > round(best_val_acc, 2) or (val_acc_rounded == round(best_val_acc, 2) and val_loss < best_val_loss):
              best_val_acc = val_accuracy
              best_val_loss = val_loss
              best_model_path = checkpoint_epoch_path

    if val_loader is not None:
      # End training time tracking
      end_time = time.time()
      elapsed_time = end_time - start_time
      print(f"Finished Training. Total time elapsed: {elapsed_time:.2f} seconds")
      print(f"Best model checkpoint: {best_model_path} with Validation Accuracy: {best_val_acc* 100:.4f}%, Validation Loss: {best_val_loss:.4f}")

      # Save training and validation metrics to CSV files
      epochs = np.arange(1, num_epochs + 1)
      np.savetxt(f"{model_path}_train_loss.csv", train_loss_list)
      np.savetxt(f"{model_path}_train_acc.csv", train_acc_list)
      np.savetxt(f"{model_path}_val_loss.csv", val_loss_list)
      np.savetxt(f"{model_path}_val_acc.csv", val_acc_list)

    return model
````

````python
import os
import shutil
from collections import defaultdict
from google.colab import files
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image


# Custom dataset class for loading images and labels
class CustomDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        """
        Initialize the dataset with image file paths and labels.

        Args:
        - image_paths (list): List of image file paths.
        - labels (list): Corresponding labels for the images.
        - transform (callable, optional): Transformations to apply to the images.
        """
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

        # Create a mapping from label names to indices
        self.label_to_idx = {label: idx for idx, label in enumerate(sorted(set(labels)))}
        self.idx_to_label = {idx: label for label, idx in self.label_to_idx.items()}

    def __len__(self):
        """Returns the total number of images in the dataset."""
        return len(self.image_paths)

    def __getitem__(self, idx):
        """
        Retrieve an image and its corresponding label.

        Args:
        - idx (int): Index of the image.

        Returns:
        - image (Tensor): Transformed image tensor.
        - label (int): Corresponding label index.
        """
        image_path = self.image_paths[idx]
        label = self.labels[idx]
        image = Image.open(image_path).convert("RGB")  # Convert to RGB format

        if self.transform:
            image = self.transform(image)

        return image, self.label_to_idx[label]

# Function to create a DataLoader for sanity check
def get_sanity_check_loader(data_dir, batch_size):
    """
    Loads images, applies transformations, and returns a DataLoader.

    Args:
    - data_dir (str): Directory containing image files.
    - batch_size (int): Number of images per batch.

    Returns:
    - train_loader (DataLoader): DataLoader for training images.
    - None: Placeholder for validation DataLoader.
    - None: Placeholder for test DataLoader.
    - classes (set): Set of unique labels found in the dataset.
    """

    # Define transformations for images
    transform = transforms.Compose([
        transforms.Resize((224, 224)),  # Resize images to match CNN input requirements
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalize to [-1, 1] range
    ])

    # Initialize lists and dictionaries for storing file information
    image_paths = []  # List of image file paths
    labels = []  # List of labels
    classes = set()  # Set of unique class labels
    person_to_letters = defaultdict(lambda: defaultdict(list))  # Dictionary to store person-to-letter mappings

    # Iterate through all files in the directory
    for filename in os.listdir(data_dir):
        if filename.endswith(".jpg"):  # Process only JPG images
            parts = filename.split("_")
            if len(parts) == 3:  # Ensure the filename follows the expected format
                person_id, letter, _ = parts  # Extract person ID and letter (label)
                image_path = os.path.join(data_dir, filename)

                # Store mappings for tracking occurrences of letters for each person
                person_to_letters[person_id][letter].append(len(image_paths))

                # Append the image path and label
                image_paths.append(image_path)
                labels.append(letter)
                classes.add(letter)  # Add the label to the set of classes

    # Create a dataset and DataLoader
    dataset = CustomDataset(image_paths, labels, transform=transform)
    train_loader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

    return train_loader, None, None, classes  # Return the train loader and placeholders for validation/test sets
````

````python
import torch
import torch.nn as nn
import torch.nn.functional as F

class newGestureCNN(nn.Module):
    def __init__(self, conv1_channels=128, fc1_units=128, num_classes=9, kernel=5, stride=1, padding=2):
        super(newGestureCNN, self).__init__()
        self.name = "newGestureCNN"

        # Convolutional layer (input shape: [N, 256, 6, 6])
        self.conv1 = nn.Conv2d(256, conv1_channels, kernel, stride=stride, padding=padding)

        # Adaptive Pooling ensures fixed output size (3x3)
        self.adaptive_pool = nn.AdaptiveAvgPool2d((3, 3))

        # Fully connected layers
        self.fc1 = nn.Linear(conv1_channels * 3 * 3, fc1_units)
        self.fc2 = nn.Linear(fc1_units, num_classes)

    def forward(self, x):
        x = F.relu(self.conv1(x))  # Convolution + ReLU
        x = self.adaptive_pool(x)  # Adaptive pooling ensures (N, 128, 3, 3)
        x = x.view(x.size(0), -1)  # Flatten feature map
        x = F.relu(self.fc1(x))  # Fully connected layer with ReLU activation
        x = self.fc2(x)  # Output layer

        return x

# Example usage
num_classes = 9
model = newGestureCNN(num_classes=num_classes, padding=0)  # Now works with any padding

# Simulated input feature tensor from AlexNet (batch size 32)
features = torch.randn(32, 256, 6, 6)  # Shape: [N, 256, 6, 6]

# Forward pass
output = model(features)  # Shape: [N, num_classes]
prob = F.softmax(output, dim=1)  # Convert logits to probabilities

print("Output shape:", output.shape)  # Should be [32, 9]
````

Train your new network, including any hyperparameter tuning. Plot and submit the training curve of your

Note: Depending on how you are caching (saving) your AlexNet features, PyTorch might still be tracking
updates to the **AlexNet weights**, which we are not tuning. One workaround is to convert your AlexNet
feature tensor into a numpy array, and then back into a PyTorch tensor.

````python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from itertools import product  # For grid search

# Load precomputed features
train_data = torch.load("saved_features/train_features.pth")
val_data = torch.load("saved_features/val_features.pth")

train_features, train_labels = train_data["features"], train_data["labels"]
val_features, val_labels = val_data["features"], val_data["labels"]

# Convert features to PyTorch tensors explicitly (remove gradient tracking)
train_features = torch.from_numpy(train_features.detach().numpy()).float()
val_features = torch.from_numpy(val_features.detach().numpy()).float()

# Move to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
train_features, train_labels = train_features.to(device), train_labels.to(device)
val_features, val_labels = val_features.to(device), val_labels.to(device)

# Define hyperparameter search space
conv1_channels_values = [64, 128, 256]
kernel_size_values = [3, 5, 7]
learning_rate_values = [0.001, 0.0005, 0.0001]
num_epochs = 20

best_val_loss = float("inf")  # Track best validation loss
best_model = None
best_hyperparams = None

# Grid Search
for conv1_channels, kernel_size, lr in product(conv1_channels_values, kernel_size_values, learning_rate_values):
    print(f"Training with conv1_channels={conv1_channels}, kernel_size={kernel_size}, lr={lr}")

    # Define model with current hyperparameters
    model = newGestureCNN(conv1_channels=conv1_channels, kernel=kernel_size, num_classes=9).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    train_losses = []
    val_losses = []

    # Training loop
    for epoch in range(num_epochs):
        model.train()
        optimizer.zero_grad()

        # Forward pass
        outputs = model(train_features)
        loss = criterion(outputs, train_labels)

        # Backward pass
        loss.backward()
        optimizer.step()

        # Validation step
        model.eval()
        with torch.no_grad():
            val_outputs = model(val_features)
            val_loss = criterion(val_outputs, val_labels)

        # Store loss values
        train_losses.append(loss.item())
        val_losses.append(val_loss.item())

        # Print progress
        print(f"Epoch [{epoch+1}/{num_epochs}] - Loss: {loss.item():.4f} - Val Loss: {val_loss.item():.4f}")

    # Store the best model based on validation loss
    avg_val_loss = sum(val_losses) / len(val_losses)
    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        best_model = model.state_dict()
        best_hyperparams = (conv1_channels, kernel_size, lr)

        # Save the best model so far
        torch.save(best_model, "best_gesture_cnn.pth")

# Print the best hyperparameters
print(" Best Model Found:")
print(f"  - conv1_channels: {best_hyperparams[0]}")
print(f"  - kernel_size: {best_hyperparams[1]}")
print(f"  - learning_rate: {best_hyperparams[2]}")
print(f"  - Best Validation Loss: {best_val_loss:.4f}")

# Plot training curves for the best model
plt.plot(range(1, num_epochs + 1), train_losses, label="Train Loss")
plt.plot(range(1, num_epochs + 1), val_losses, label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.title(f"Training Curve (Best Model: {best_hyperparams})")
plt.show()

print("Training complete. Best model saved as 'best_gesture_cnn.pth'.")
````

````python
import torch
import torch.nn as nn
import torch.nn.functional as F

class newGestureCNN18(nn.Module):
    def __init__(self,
                 input_channels=512,  # Matches the feature extractor output
                 conv1_channels=128,  # Number of filters in the convolution layer
                 kernel_size=3,       # Flexible kernel size
                 stride=1,            # Flexible stride
                 auto_padding=True,   # Automatically adjust padding
                 fc1_units=128,       # Fully connected layer size
                 num_classes=9,       # Number of output classes
                 input_size=(1, 1)):  # Input feature map size from backbone
        super(newGestureCNN18, self).__init__()
        self.name = "newGestureCNN18"

        # Auto-adjust padding to maintain spatial size
        if auto_padding:
            padding = (kernel_size - 1) // 2  # Same padding formula
        else:
            padding = 0  # No padding if disabled

        # First convolutional layer
        self.conv1 = nn.Conv2d(input_channels, conv1_channels,
                               kernel_size=kernel_size, stride=stride, padding=padding)

        # Adaptive Pooling ensures consistent output size
        self.adaptive_pool = nn.AdaptiveAvgPool2d((1, 1))  # Reduces spatial size to (1,1)

        # Fully connected layers
        self.fc1 = nn.Linear(conv1_channels * 1 * 1, fc1_units)
        self.fc2 = nn.Linear(fc1_units, num_classes)

    def forward(self, x):
        x = F.relu(self.conv1(x))  # Apply convolution + ReLU
        x = self.adaptive_pool(x)  # Adaptive pooling ensures (N, conv1_channels, 1, 1)
        x = x.view(x.size(0), -1)  # Flatten feature map
        x = F.relu(self.fc1(x))  # Fully connected layer with ReLU
        x = self.fc2(x)  # Final output layer (logits)
        return x

# Works for ResNet18 Features (512,1,1)
resnet_model = newGestureCNN18(input_channels=512, kernel_size=3, stride=1, auto_padding=True)
resnet_features = torch.randn(32, 512, 1, 1)
output_resnet = resnet_model(resnet_features)
print("ResNet18 Output Shape:", output_resnet.shape)  # Expected: [32, 9]
````

````python
import os
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
from collections import defaultdict

# Custom dataset class
class CustomDataset(Dataset):
    def __init__(self, image_dir, transform=None):
        """
        Initializes the dataset with image file paths and labels.

        Args:
        - image_dir (str): Directory containing image files.
        - transform (callable, optional): Transformations to apply to the images.
        """
        self.image_dir = image_dir
        self.transform = transform
        self.image_paths = []
        self.labels = []
        self.classes = set()  # Store unique classes


        # Iterate through the directory to collect image paths and labels
        for filename in os.listdir(image_dir):
            if filename.endswith(".jpg"):  # Ensure it's an image file
                parts = filename.split("_")
                if len(parts) == 3:  # Check for correct filename format
                    class_label = parts[1]  # Extract class label from filename
                    image_path = os.path.join(image_dir, filename)
                    self.image_paths.append(image_path)
                    self.labels.append(class_label)
                    self.classes.add(class_label)
        # Convert set to a sorted list so classes start from 'A'
        self.classes = sorted(list(self.classes))
        # Create label-to-index mapping
        self.label_to_idx = {label: idx for idx, label in enumerate(sorted(self.classes))}
        self.idx_to_label = {idx: label for label, idx in self.label_to_idx.items()}

    def __len__(self):
        """Returns the total number of images in the dataset."""
        return len(self.image_paths)

    def __getitem__(self, idx):
        """
        Retrieves an image and its corresponding label.

        Args:
        - idx (int): Index of the image.

        Returns:
        - image (Tensor): Transformed image tensor.
        - label (int): Corresponding label index.
        """
        image_path = self.image_paths[idx]
        label = self.labels[idx]
        image = Image.open(image_path).convert("RGB")  # Load and convert the image to RGB

        if self.transform:
            image = self.transform(image)

        return image, self.label_to_idx[label]  # Return image tensor and label index


# Define transformations for images
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize images to match AlexNet input requirements
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalize to [-1, 1] range
])

# Create DataLoader
batch_size = 27  # Total 27 images
dataset = CustomDataset(image_dir="sample", transform=transform)
data_loader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

# Print class index mapping
classes = dataset.classes

import torchvision.models as models

# Load pre-trained AlexNet model
alexnet = models.alexnet(pretrained=True)
alexnet.eval()  # Set to evaluation mode

# Save extracted features
extract_and_save_features(data_loader, "saved_features_test/test_features.pth")
````

````python
# Task 2

import torch
import random
import numpy as np
import os
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset
from collections import defaultdict
from PIL import Image
import glob
import matplotlib.pyplot as plt

# Set random seed for reproducibility
torch.manual_seed(999)
random.seed(999)
np.random.seed(999)

# Load texture backgrounds
texture_paths = glob.glob("textures/*.jpg")  # Folder where texture images are stored
textures = [Image.open(path).convert("RGB") for path in texture_paths]

def apply_texture_background(img):
    """Replaces the image's background with a random texture."""
    texture = random.choice(textures)
    texture = texture.resize(img.size)
    img = Image.blend(img, texture, alpha=0.2)  # Blend the image with the texture
    return img

def data_augmentation_loader(data_dir, batch_size, valid_split=0.2, test_split=0.1):
    """
    Loads dataset while ensuring unique individuals across splits, and applies data augmentation
    by replacing clean backgrounds with textured backgrounds.
    """

    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.Lambda(lambda img: apply_texture_background(img)),  # Apply texture replacement
        transforms.RandomRotation(10),
        transforms.ColorJitter(brightness=0.2, contrast=0.2),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
    ])

    dataset = datasets.ImageFolder(data_dir, transform=transform)
    classes = dataset.classes  # List of class names (A-I)

    person_to_letters = defaultdict(lambda: defaultdict(list))
    for idx, (img_path, label) in enumerate(dataset.imgs):
        filename = os.path.basename(img_path)
        parts = filename.split("_")
        if len(parts) == 3:
            person_id, letter, seq = parts
            person_to_letters[person_id][letter].append(idx)

    valid_people = {person: images for person, images in person_to_letters.items() if len(images) == 9 and all(len(images[letter]) == 3 for letter in images)}
    invalid_people = {person: images for person, images in person_to_letters.items() if person not in valid_people}

    total_people = len(person_to_letters)
    valid_people_list = list(valid_people.keys())
    random.shuffle(valid_people_list)

    test_people_count = round(test_split * total_people)
    remaining_valid_count = len(valid_people_list) - test_people_count
    valid_people_count = round((total_people - test_people_count) * valid_split)

    test_people = set(valid_people_list[:test_people_count])
    train_valid_people = valid_people_list[test_people_count:]
    val_people = set(train_valid_people[:valid_people_count])
    train_people = set(train_valid_people[valid_people_count:])
    train_people.update(invalid_people.keys())

    test_indices = [idx for person in test_people for letter in valid_people[person].values() for idx in letter]
    val_indices = [idx for person in val_people for letter in valid_people[person].values() for idx in letter]
    train_indices = [idx for person in train_people for letter in person_to_letters[person].values() for idx in letter]

    train_loader = DataLoader(Subset(dataset, train_indices), batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(Subset(dataset, val_indices), batch_size=batch_size, shuffle=False)
    test_loader = DataLoader(Subset(dataset, test_indices), batch_size=batch_size, shuffle=False)

    return train_loader, val_loader, test_loader, classes

def show_augmented_images(data_loader, classes):
    """Displays a batch of augmented images."""
    images, labels = next(iter(data_loader))
    fig, axes = plt.subplots(4, 8, figsize=(12, 6))
    axes = axes.flatten()
    for img, label, ax in zip(images, labels, axes):
        img = img.permute(1, 2, 0).numpy() * 0.5 + 0.5  # Convert back to RGB
        ax.imshow(img)
        ax.set_title(classes[label])
        ax.axis("off")
    plt.show()

# Load augmented dataset
train_loader, val_loader, test_loader, classes = data_augmentation_loader("dataset", batch_size=32)
show_augmented_images(train_loader, classes)

````

````python
# Task 1

import torch
import torchvision.models as models
import os
import torch.nn as nn
import torch.nn.functional as F

class newGestureCNN(nn.Module):
    def __init__(self, conv1_channels=128, fc1_units=128, num_classes=9, kernel=5, stride=1, padding=2):
        super(newGestureCNN, self).__init__()
        self.name = "newGestureCNN"

        # Convolutional layer (input shape: [N, 256, 6, 6])
        self.conv1 = nn.Conv2d(256, conv1_channels, kernel, stride=stride, padding=padding)

        # Adaptive Pooling ensures fixed output size (3x3)
        self.adaptive_pool = nn.AdaptiveAvgPool2d((3, 3))

        # Fully connected layers
        self.fc1 = nn.Linear(conv1_channels * 3 * 3, fc1_units)
        self.fc2 = nn.Linear(fc1_units, num_classes)

    def forward(self, x):
        x = F.relu(self.conv1(x))  # Convolution + ReLU
        x = self.adaptive_pool(x)  # Adaptive pooling ensures (N, 128, 3, 3)
        x = x.view(x.size(0), -1)  # Flatten feature map
        x = F.relu(self.fc1(x))  # Fully connected layer with ReLU activation
        x = self.fc2(x)  # Output layer

        return x

# Define device (use GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load Pretrained AlexNet
alexnet = models.alexnet(pretrained=True).to(device)
alexnet.eval()  # Set to evaluation mode

# Define save directory for extracted features
save_dir = "saved_features"
os.makedirs(save_dir, exist_ok=True)

# Extract and save features for training, validation, and test sets
# extract_and_save_features(test_loader, os.path.join(save_dir, "test_features.pth"))

# Load extracted test dataset
test_data = torch.load(os.path.join(save_dir, "test_features.pth"))
test_features, test_labels = test_data["features"], test_data["labels"]

# Move data to GPU if available
test_features, test_labels = test_features.to(device), test_labels.to(device)

# Define Model with Current Hyperparameters
conv1_channels = 256
kernel = 5

# Initialize Model
model = newGestureCNN(conv1_channels=conv1_channels, kernel=kernel, num_classes=9).to(device)

# Load trained model weights
model_path = "best_gesture_cnn.pth"
model.load_state_dict(torch.load(model_path, map_location=device))  # Load state dict properly
model.eval()  # Set to evaluation mode

import torch
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

def evaluate_model(model, test_features, test_labels, classes):
    """Evaluates model performance using extracted features and plots a confusion matrix."""
    model.to(device)
    correct = 0
    total = test_labels.size(0)

    with torch.no_grad():  # Disable gradient calculation for efficiency
        outputs = model(test_features)  # Forward pass
        predictions = torch.argmax(outputs, dim=1)  # Get predicted class

        correct += (predictions == test_labels).sum().item()

    accuracy = correct / total * 100
    print(f"Model Accuracy on Test Set: {accuracy:.2f}% with data augmentation images.")

    # Compute confusion matrix
    conf_matrix = confusion_matrix(test_labels.cpu().numpy(), predictions.cpu().numpy())

    # Normalize the confusion matrix
    conf_matrix_normalized = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:, np.newaxis]

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix_normalized, annot=True, cmap="Blues", fmt=".2f",
                xticklabels=classes, yticklabels=classes)
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.title("Normalized Confusion Matrix")
    plt.show()

# Call the function with class labels
evaluate_model(model, test_features, test_labels, classes)
````
